## Making the Graph

I started by creating a graph and adding nodes. Each node was assigned to one of the names that are in the assignment sample. I added edges to connect all of the nodes, the same way that they are connected in the sample, making sure to double check that the lines matched up. Then, I tried to match the gray icons and the purple line as well, making sure to add the name labels too.  I then printed the graph with the 'nx.draw' function

```python
g = nx.Graph()

nodes = ['Jane', 'Ike', 'Heather', 'Fernando','Garth','Diane','Carol','Andre','Beverly','Ed']
g.add_nodes_from(nodes)

g.add_edge('Jane','Ike')
g.add_edge('Ike','Heather')

g.add_edge('Heather','Fernando')
g.add_edge('Heather','Garth')

g.add_edge('Garth','Fernando')
g.add_edge('Garth','Diane')
g.add_edge('Fernando','Diane')
g.add_edge('Fernando','Andre')
g.add_edge('Garth','Beverly')


g.add_edge('Fernando','Carol')
g.add_edge('Carol','Diane')

g.add_edge('Carol','Andre')
g.add_edge('Diane','Andre')

g.add_edge('Andre','Beverly')
g.add_edge('Beverly','Diane')

g.add_edge('Beverly','Ed')
g.add_edge('Garth','Ed')
g.add_edge('Beverly','Diane')
g.add_edge('Ed','Diane')

nx.draw(g, with_labels=True, font_size=8, font_color='black', font_weight='bold', node_size=700, node_color='lightgray',edge_color='purple')

```
